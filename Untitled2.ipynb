```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
G.add_nodes_from([2,3])
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(1,3)
G.add_edge(3,1)
G.add_nodes_from([4,5])
G.add_edge(5,1)
G.add_nodes_from([6,7,8])
G.add_edges_from([(6,7), (7,8)])
G.add_edge(21,22)
nx.draw(G)
#plt.show()

print(G.number_of_edges())
G.number_of_nodes()
G.edges
G.edges
#G.graph
from networkx.readwrite import json_graph

json_data=json_graph.node_link_data(G)
print(json_data)
json_data_to_graph=json_graph.node_link_graph(json_data)
print(json_data_to_graph)

G.add_nodes_from([9,10,11])
```

```python
G.is_directed()
```

```python
for line in nx.generate_graphml(G):
    print(line)
```
